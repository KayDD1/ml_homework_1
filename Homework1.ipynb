```python
df_island = df[df.ocean_proximity == 'ISLAND'][['housing_median_age', 'total_rooms', 'total_bedrooms']]
```

```python
X = df_island.values
X
```

```python
XT = X.T
```

```python
XTX = XT.dot(X)
```

```python
X_inv = np.linalg.inv(XTX)
```

```python
y = np.array([950, 1300, 800, 1000, 1300])
```

```python
w = X_inv.dot(XT).dot(y)
w
```
